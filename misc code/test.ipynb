```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Create some sample data
fig, ax = plt.figure(figsize=(4, 6)), plt.gca()
SCALE = 5
# Create a line
ax.plot([0, 0], [-1, 1], color=plt.get_cmap("Set1")(8), zorder=0, label="Garis OA")

point_O = [0, 0.5]
point_P = [0.5,0]
ax.scatter(*point_O, marker='o', color=plt.get_cmap("Set1")(1), s=50, zorder=3, label="Titik O")
ax.scatter(*point_P, marker='o', color=plt.get_cmap("Set1")(2), s=50, zorder=3, label="Titik P")

vektor_OP = np.array(point_P) - np.array(point_O)
ax.quiver(*point_O, *vektor_OP, angles='xy', scale=SCALE/2, color=plt.get_cmap("Set1")(3), zorder=2, label=r"$\overrightarrow{OP}$")


ax.quiver(*point_O, *[0,1], scale=SCALE, color=plt.get_cmap("Set1")(4),  label=r"$\hat{n}$")

point_P_ = [2*point_P[0],point_P[1]]
ax.scatter(*point_P_, alpha = 0)
ax.scatter(*[-0.05,0], alpha = 0)


# Add the "out of the page" symbol (⊙) at a specific point
vektor_point = point_O
# ax.text(
#     vektor_point[0],
#     vektor_point[1],
#     "⊙",
#     fontsize=24,
#     ha="center",
#     va="center",
#     color=plt.get_cmap("Set1")(0),
# )
# Add the "in the page" symbol (⊗) at a specific point
ax.text(
    vektor_point[0],
    vektor_point[1],
    "⊗",
    fontsize=24,
    ha="center",
    va="center",
    color=plt.get_cmap("Set1")(0),
)
ax.text(
    vektor_point[0]+0.05,
    vektor_point[1],
    r"Hasil $\hat{n} \times \overrightarrow{OP}$",
    fontsize=12,
    ha="left",
    va="bottom",
    color=plt.get_cmap("Set1")(0),
)


vektor_point = [0, -0.5]
ax.text(
    vektor_point[0],
    vektor_point[1],
    "⊙",
    fontsize=24,
    ha="center",
    va="center",
    color=plt.get_cmap("Set1")(6),
)
ax.text(
    vektor_point[0]+0.05,
    vektor_point[1],
    r"Arah normal bidang $XY$",
    fontsize=12,
    ha="left",
    va="center",
    color=plt.get_cmap("Set1")(6),
)

# Set the aspect of the plot to be equal
# ax.set_aspect("equal", adjustable="datalim")
# Display the plot
ax.legend()
fig.tight_layout()

# Set figure size
# fig.set_size_inches(3, 6)

# Save the plot as SVG file
plt.savefig("titik P di kanan.svg")
plt.show()
```
